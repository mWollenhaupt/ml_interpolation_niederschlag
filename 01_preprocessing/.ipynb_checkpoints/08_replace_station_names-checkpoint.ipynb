```python
import pandas as pd
import datetime
```

```python
ip_file = "prediction_names.csv"
ip_df = pd.read_csv(ip_file, delimiter=";")
obs_file = "observation.csv"
obs_df = pd.read_csv(obs_file, delimiter=";")
```

##### Name - Value dict for observation

```python
data = {}
for row in obs_df.iterrows():
    cr = row[1]
    data[cr['name']] = cr['value']
```

## 10 neighbours

```python
cc = 0
pc = 1
all = len(ip_df)
utc = str(obs_df['utc'][0])
doy = obs_df['doy'][0]
for idx, row in ip_df.iterrows():
    cc = cc+1
    per = (cc/all)*100
    
    ip_df.at[idx, 'utc'] = utc
    ip_df.at[idx, 'doy'] = doy
    ip_df.at[idx, 'value_1'] = data[row['value_1']]
    ip_df.at[idx, 'value_2'] = data[row['value_2']]
    ip_df.at[idx, 'value_3'] = data[row['value_3']]
    ip_df.at[idx, 'value_4'] = data[row['value_4']]
    ip_df.at[idx, 'value_5'] = data[row['value_5']]
    ip_df.at[idx, 'value_6'] = data[row['value_6']]
    ip_df.at[idx, 'value_7'] = data[row['value_7']]
    ip_df.at[idx, 'value_8'] = data[row['value_8']]
    ip_df.at[idx, 'value_9'] = data[row['value_9']]
    ip_df.at[idx, 'value_10'] = data[row['value_10']]
    if per > pc:
        print('{0}: {1:.1f}%'.format(datetime.datetime.now(), pc))
        pc = pc+1
```

```python
ip_df = ip_df.sort_values(['north', 'east'], ascending=[False, True])
```
