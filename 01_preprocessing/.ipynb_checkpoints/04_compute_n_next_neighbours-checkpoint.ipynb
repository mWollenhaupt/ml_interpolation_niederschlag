```python
import pandas as pd
from math import sqrt
from datetime import datetime
```

```python
file = "observations_filterd_neigbours.csv"
df = pd.read_csv(file, delimiter=";")
```

```python
groups = df.groupby('timestamp')
```

```python
def findNextNeighbours(obs, station, number=3):
    obs = obs[obs.name != station[0]]
    result = []
    east = station[1]
    north = station[2]
    up = station[3]
    for row in obs.iterrows():
        cr = row[1]
        dist = sqrt((east-cr['east'])*(east-cr['east'])+(north-cr['north'])*(north-cr['north'])+(up-cr['up'])*(up-cr['up']))
        r = [dist, cr['value']]
        result.append(r)
    result.sort(key=lambda x: x[0])
    return result[0:number]
```

```python
def prepareList(result):
    newList = []
    for k in range(len(result)):
        n = [j for i in result[k] for j in i]
        newList.append(n)
    return newList
```

```python
def createFinalDataframe(result):
    cols = []
    l = len(result[0])
    l = l/2
    for i in range(int(l)):
        cols.append('dist_{}'.format(i+1))
        cols.append('value_{}'.format(i+1))
    
    cols_df = pd.DataFrame(columns=cols, data=result)
    return cols_df
```

```python
result = []
l = len(groups)
cc = 1
pc = 1
for name, group in groups:
    perc = (cc/l)*100
    if perc > pc:
        print('{0}: {1:.1f}%'.format(datetime.now(), perc))
        pc+=1
    cc+=1
    for row in group.iterrows():
        cr = row[1]
        station = [cr['name'], cr['east'], cr['north'], cr['up']]
        result.append(findNextNeighbours(group, station, number=10))
```

```python
result = prepareList(result)
```

```python
split = createFinalDataframe(result)
```

```python
split.head()
```

```python
res = pd.concat([df, split], axis=1, sort=True)
```

```python
columns = res.columns.tolist()
print(columns)
```

```python
file = "observations_filterd_wmn_neighbours.csv"
res.to_csv(file, sep=";", columns=columns, index=False)
```
